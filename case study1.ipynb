# Plot a bar graph of male vs female buyers participated in the sales.

```python
df['Buyer Gender'].value_counts()
```

```python
h=df['Buyer Gender'].value_counts()
plt.figure(figsize=(6,6))
h.plot(kind='bar',color='green')
plt.title('Male vs. Female buyers in the sales',size=13)
plt.xlabel('Buyer Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
```
